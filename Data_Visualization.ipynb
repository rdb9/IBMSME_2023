MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

BAR PLOT

```python
x = df.head(20).Team
y = df.head(20).Goals_Against
plt.figure(figsize=(20,5))
plt.bar(x,y) 
plt.title("Bar Plot")
plt.ylabel("Goals against")
plt.xticks(rotation = 'vertical')
plt.xlabel("Team")
plt.show()
```

LINE PLOT

```python
x = df.Team
y = df.Wins
plt.figure(figsize=(20,5))
plt.plot(x,y,'-h')
plt.xlabel("Team")
plt.xticks(rotation='vertical')
plt.ylabel("Wins")
plt.title("Line Plot")
plt.show()
```

AREA CHART

```python
x = df.Team
y = df.Wins
plt.figure(figsize=(20,5))
plt.fill_between(x,y)
plt.title("Area Chart")
plt.ylabel("Wins")
plt.xticks(rotation='vertical')
plt.xlabel("Team")
plt.show()
```

HISTOGRAM

```python
plt.hist(df.head(40).Team, density = 1)
plt.xticks(rotation = 'vertical')
plt.title("Histogram")
plt.show()
```

BOX PLOT

```python
items = [df.Losses, df.Goals_Against, df.Games_Played, df.Wins]
plt.boxplot(items)
plt.title("Box Plot")
plt.show()
```

PIE CHART

```python
y = df.Team
x = df.Goals_Against
plt.figure(figsize =(30,5))
plt.pie(x, labels=y, radius = 2)
plt.show()
```

PIE CHART WITH A SLICE ( CUT PIE )

```python
x = df.Team
y = df.Goals_Against
circle = plt.Circle( (0,0), 0.4, color = 'white')
plt.figure(figsize = (50,15))
plt.pie(y, labels = x, colors = ['b','g','r','c','m','y','k','w','r','c','m','y','k','b','g','r','c','m'],wedgeprops = {'linewidth':6,'edgecolor':'white'})
plt.title("Cut Pie")
p = plt.gcf()
p.gca().add_artist(circle)
plt.show()
```

DONUT CHART

```python
x = df.Team
y = df.Goals_Against
my_circle = plt.Circle( (0,0), 0.6, color='white')
plt.figure(figsize = (50,15))
plt.pie(y, labels=x, colors=['b','g','r','c','m','y','k','c','r','c','m','y','k','b','g','r','c','m','y'])
plt.title("Donut Chart")
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

SCATTER PLOT

```python
x = df.Team
y = df.Goals_Against
plt.figure(figsize = (20,8))
plt.scatter(x,y)
plt.title("Scatter plot")
plt.ylabel("Points")
plt.xticks(rotation='vertical')
plt.xlabel("Team")
plt.show()
```

```python
import seaborn as sns
```

PIE CHART

```python
y = df.Team
x = df.Wins
plt.figure(figsize = (20,8))
palette_color = sns.color_palette('bright')
plt.pie(x, labels=y, colors=palette_color, radius = 2,autopct='%.0f%%')
plt.show()
```

DONUT CHART

```python
y = df.Team
x = df.Wins
plt.figure(figsize = (20,8))
palette_color = sns.color_palette('bright')
my_circle = plt.Circle( (0,0), 1, color='white')
plt.pie(x, labels=y, colors=palette_color,radius = 2,shadow = 'True',autopct='%.0f%%')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

VENN DIAGRAM

```python
from matplotlib_venn import venn3, venn3_circles
plt.figure(figsize = (20,8))
v=venn3(subsets = (10, 8, 22, 6,9,4,2), set_labels = ('Andrei Vasilevskiy','Sergei Bobrovsky', 'Juuse Saros'))
c=venn3_circles(subsets = (10, 8, 22, 6,9,4,2), linestyle='dashed', linewidth=1, color="grey")
c[0].set_lw(8.0)
c[0].set_ls('dotted')
c[0].set_color('skyblue')
plt.title("Venn Diagram")
plt.show()
```

DENSITY PLOT

```python
sns.set(style="darkgrid")
plt.figure(figsize = (20,8))
sns.kdeplot(df['Wins'])
plt.title("Density Plot")
plt.show()
```

VIOLIN PLOT

```python
sns.set(style="darkgrid")
plt.figure(figsize = (20,8))
sns.violinplot(x=df["Losses"], palette="Blues")
plt.title("Violin Plot")
plt.show()
```

HISTOGRAM

```python
sns.distplot( a=df["Goals_Against"], hist=True, kde=False, rug=False )
plt.title("Histogram")
plt.show()
```

BOX PLOT

```python
sns.set_style("whitegrid")
sns.boxplot('Losses', data = df)
plt.title("Box Plot")
plt.show()
```
